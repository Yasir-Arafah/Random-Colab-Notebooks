```python
# Step 2: Import necessary libraries
from PIL import Image
import pytesseract
from google.colab import files
import pandas as pd

# Initialize an empty DataFrame to store the results
df = pd.DataFrame(columns=["Image Name", "Extracted Text", "Image Path"])
```

```python
# Function to extract text from an image
def extract_text_from_image(image_path):
    # Load the image from the specified file path
    image = Image.open(image_path)

    # Use pytesseract to extract text from the image
    text = pytesseract.image_to_string(image)

    return text

# Step 3: Upload images
uploaded = files.upload()

# Step 4: Extract text and save the image path for each uploaded image
for image_name in uploaded.keys():
    # Extract text from the image
    extracted_text = extract_text_from_image(image_name)

    # Create a new row with the data
    new_row = pd.DataFrame({
        "Image Name": [image_name],
        "Extracted Text": [extracted_text],
        "Image Path": [image_name]
    })

    # Concatenate the new row to the DataFrame
    df = pd.concat([df, new_row], ignore_index=True)

# Display the DataFrame
df
```

```python
# Step 2: Import necessary libraries
from PIL import Image
import pytesseract
from google.colab import files
import pandas as pd
import os
from google.colab import drive

# Initialize an empty DataFrame to store the results
df = pd.DataFrame(columns=["Image Name", "Extracted Text", "Image Path"])
```

```python
# Specify the folder where you want to save the images
drive_folder = '/content/drive/MyDrive/ExtractedImages'  # Replace with your desired folder path in Google Drive

# Ensure the directory exists
os.makedirs(drive_folder, exist_ok=True)

# Function to extract text from an image
def extract_text_from_image(image_path):
    # Load the image from the specified file path
    image = Image.open(image_path)

    # Use pytesseract to extract text from the image
    text = pytesseract.image_to_string(image)

    return text

# Step 4: Upload images
uploaded = files.upload()

# Step 5: Extract text and save the image in the Drive folder
for image_name in uploaded.keys():
    # Extract text from the image
    extracted_text = extract_text_from_image(image_name)

    # Save the image to the specified Google Drive folder
    image_save_path = os.path.join(drive_folder, image_name)
    with open(image_save_path, 'wb') as f:
        f.write(uploaded[image_name])

    # Create a new row with the data
    new_row = pd.DataFrame({
        "Image Name": [image_name],
        "Extracted Text": [extracted_text],
        "Image Path": [image_save_path]
    })

    # Concatenate the new row to the DataFrame
    df = pd.concat([df, new_row], ignore_index=True)

# Display the DataFrame
df
```
